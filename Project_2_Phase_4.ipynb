## Section II: Relationship Visualizations

Exploring relationships between numerical variables using a scatter plot and correlation heatmap.

```python
# Scatter Plot – Experience vs Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_years', y='salary', data=df, alpha=0.1, s=10, color='teal')
plt.title('Experience Years vs Salary', fontsize=14)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.tight_layout()
plt.show()
```

### Scatter Plot Observation

The scatter plot reveals a **moderate positive relationship** between years of experience and salary (correlation ≈ 0.44). As experience increases, salary generally trends upward. However, there is significant spread at every experience level, meaning experience alone does not fully determine salary. Other factors like education, skills count, and industry likely contribute to the variation seen at each level of experience.

```python
# Correlation Heatmap – All Numerical Variables
plt.figure(figsize=(8, 6))
corr_matrix = df.select_dtypes(include='number').corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numerical Variables', fontsize=14)
plt.tight_layout()
plt.show()
```

### Heatmap Observation

The correlation heatmap shows the relationships between all numerical variables in the dataset:

- **Experience years & salary (0.44):** The strongest correlation. More experience is moderately associated with higher pay.
- **Skills count & salary (0.13):** A weak positive correlation, suggesting that having more skills provides a slight salary advantage.
- **Certifications & salary (0.07):** Nearly negligible correlation, indicating certifications alone do not significantly impact salary in this dataset.
- **Between predictors:** Experience, skills count, and certifications show virtually no correlation with each other (all ≈ 0.00), meaning these features are independent.

Overall, experience is the most meaningful numerical predictor of salary, while certifications and skills count have minimal linear impact. This suggests that non-numerical factors (industry, education level, company size) or non-linear relationships may also play important roles.
