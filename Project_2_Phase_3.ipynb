Individual & Comparative Visualization 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("job_salary_prediction_dataset.csv")
df.head()
```

```python
df.columns
```

Histogram \(Numerical Column\)


```python
# Get histogram data without plotting
counts, bins = np.histogram(df["salary"], bins=20)

# Create colors (one per bin)
colors = plt.cm.viridis(np.linspace(0, 1, len(counts)))

# Plot bars manually
plt.figure()
plt.bar(bins[:-1], counts, width=np.diff(bins), color=colors, edgecolor='black', align='edge')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

This histogram shows how many people earn different amounts of money. It helps us see if most people earn similar amounts or if many people earn either a lot more or a lot less. If the graph leans to one side, it means most people earn either lower or higher salaries.


Visualization 2: Bar Chart \(Categorical Column\)


```python
df["job_title"].value_counts().plot(kind="bar", color=colors)

plt.title("Count of Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.show()
```

This bar chart shows how many people have each job. It helps find out which jobs are the most common.


Count Plot


```python
sns.countplot(x="education_level", data=df)

plt.title("Education Level Distribution")
plt.xlabel("Education Level")
plt.ylabel("Count")

plt.xticks(rotation=45)
plt.show()
```

COMPARATIVE VISUALIZATION


Average Salary by Job Title


```python
avg_salary = df.groupby("job_title")["salary"].mean()

avg_salary.plot(kind="bar", color=colors)

plt.title("Average Salary by Job Title")
plt.xlabel("Job Title")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.show()
```

This bar chart shows how much people earn in different jobs. It helps us see which jobs pay more and which pay less. It shows the differences in pay between the jobs.The chart shows that some tech jobs pay more than others. AI and Machine Learning Engineers earn the most, around \$175,000. Jobs like Cloud, DevOps, and Backend Engineering usually pay between \$145,000 and \$155,000. Data and Business Analysts earn a bit less, between \$120,000 and \$125,000. This means that jobs in new technologies and product management often pay more than more basic roles like analysis or front\-end work.
