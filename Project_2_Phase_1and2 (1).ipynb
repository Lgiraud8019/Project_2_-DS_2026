### TASK 06 :: ADD A NEW COLUMN

```python
import pandas as pd

job_df = pd.read_csv('job_salary_prediction_dataset.csv')
```

```python
job_df['salary_per_skill'] = job_df['salary'] / job_df['skills_count']
```

```python
job_df.head()
```

##### The salary_per_skill column represents the average amount of salary associated with each skill required for a given job. It is calculated by dividing the total salary by the number of skills listed for that position, giving a rough estimate of how much each individual skill contributes to overall pay.

### TASK 07 :: FILTER THE DATA

```python
high_paid_remote = job_df[(job_df['salary'] > 100000) & (job_df['remote_work'] == 1)]
high_paid_remote.head()
```

```python
experienced_grads = job_df[(job_df['experience_years'] >= 5) & (job_df['education_level'] == 'Bachelor')]
experienced_grads.head()
```

##### Two subsets were created to isolate meaningful patterns within the dataset based on salary, work style, experience, and education.

The first subset high_paid_remote includes jobs that offer salaries above $100,000 and are classified as remote positions. This subset is interesting because it highlights high-paying opportunities that also provide flexibility in scheduling.

The second subset experienced_grads focuses on jobs that require at least five years of experience and a Bachelor’s degree. This subset is useful for understanding mid- to senior-level positions and the typical expectations tied to them. It allows you to explore how experience and education together influence job qualifications and salary potential.

Together, these subsets are interesting because they narrow down the dataset into targeted groups, making it easier to identify patterns in them and draw conclusions about high-value jobs.
